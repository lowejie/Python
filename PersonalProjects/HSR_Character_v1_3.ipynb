Segregation of characters based on defense value at level 80 by percentiles

```python
q25, q50, q75, q100 = hsr_df['def_80'].quantile([0.25, 0.5, 0.75, 1])

hsr_df_sorted_def80 = hsr_df.sort_values('def_80', ascending=False)

hsr_df_def80_25 = hsr_df_sorted_def80[['character','def_80','path']][hsr_df['def_80'] <= q25]
hsr_df_def80_50 = hsr_df_sorted_def80[['character','def_80','path']][(hsr_df['def_80'] > q25) & (hsr_df['def_80'] <= q50)]
hsr_df_def80_75 = hsr_df_sorted_def80[['character','def_80','path']][(hsr_df['def_80'] > q50) & (hsr_df['def_80'] <= q75)]
hsr_df_def80_100 = hsr_df_sorted_def80[['character','def_80','path']][hsr_df['def_80'] > q75]

print("25th percentile data:\n", hsr_df_def80_25)
print("50th percentile data:\n", hsr_df_def80_50)
print("75th percentile data:\n", hsr_df_def80_75)
print("100th percentile data:\n", hsr_df_def80_100)
```

Define assign_score function to assign values 1 to 4 based on percentile.

```python
def assign_score(value):
    if value <= q25:
        return 1
    elif value <= q50:
        return 2
    elif value <= q75:
        return 3
    else:
        return 4
```

Apply assign_score function to add new defense score column to original dataframe.

```python
hsr_df['def_score'] = hsr_df['def_80'].apply(assign_score)

hsr_df
```

Segregation done similar for hp value at level 80

```python
q25, q50, q75, q100 = hsr_df['hp_80'].quantile([0.25, 0.5, 0.75, 1])

hsr_df_sorted_hp80 = hsr_df.sort_values('hp_80', ascending=False)

hsr_df_hp80_25 = hsr_df_sorted_hp80[['character','hp_80','path']][hsr_df['hp_80'] <= q25]
hsr_df_hp80_50 = hsr_df_sorted_hp80[['character','hp_80','path']][(hsr_df['hp_80'] > q25) & (hsr_df['hp_80'] <= q50)]
hsr_df_hp80_75 = hsr_df_sorted_hp80[['character','hp_80','path']][(hsr_df['hp_80'] > q50) & (hsr_df['hp_80'] <= q75)]
hsr_df_hp80_100 = hsr_df_sorted_hp80[['character','hp_80','path']][hsr_df['hp_80'] > q75]

print("25th percentile data:\n", hsr_df_hp80_25)
print("50th percentile data:\n", hsr_df_hp80_50)
print("75th percentile data:\n", hsr_df_hp80_75)
print("100th percentile data:\n", hsr_df_hp80_100)
```

Assign score function used to create hp score column on original dataframe.

```python
hsr_df['hp_score'] = hsr_df['hp_80'].apply(assign_score)

hsr_df
```

Display characters that have value 4 for both def_score and hp_score, implying that they are more "tanky".

```python
hsr_df[(hsr_df['def_score'] == 4) & (hsr_df['hp_score'] == 4)]
```

```python
hsr_tankiness_df = hsr_df.sort_values(['def_score','hp_score'], ascending=False)

hsr_tankiness_df
```

Write results back into CSV file

```python
hsr_tankiness_df = hsr_tankiness_df[[
    'character',
    'rarity',
    'path',
    'combat_type',
    'atk_80',
    'def_80',
    'hp_80',
    'def_score',
    'hp_score'
]]
```
